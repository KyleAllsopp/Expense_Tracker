# Read Data from CSV with Error Handling

Here we are defining a function load_from_csv that reads data from a CSV file and converts it into a list of dictionaries. We're using Python's built-in csv module. csv.DictReader reads each row of the CSV file into a dictionary where the keys are the fieldnames from the first row in the file. We also handle potential errors such as the file not existing.

```python
# Again ensuring the csv library is included
import csv

# Define the load from csv variable which converts the data inside the csv to dictionaries we can interect with here in python
def load_from_csv(filename):
    try:
        # Empty expenses dictionary
        expenses = []
        with open(filename, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                # Convert 'Amount' from string to float for better handling later
                row["Amount"] = float(row["Amount"])
                expenses.append(row)
        return expenses
    except FileNotFoundError:
        print(f"File '{filename}' not found. Please check the file name and try again.")
        return []  # Return an empty list if file is not found

# Define the expenses from the load from csv variable above looking specifically at the dummy expenses column
expenses = load_from_csv('dummy_expenses.csv')
```

# Add New Expense Entries with Data Validation

In the add_expense function, we're adding a new expense entry. This function prompts the user for input, validates it, and then appends the new entry to the existing expenses list. We're checking if the entered amount is a valid float and if the date is in the correct format (YYYY-MM-DD).

```python
# Define variable to check if the date is valid
def is_valid_date(date_str):
    try:
        year, month, day = map(int, date_str.split('-'))
        return 1 <= month <= 12 and 1 <= day <= 31  # Simplified check
    except ValueError:
        return False

# Define variable for adding a new expense 
def add_expense(expenses):
    while True:  # Start a loop that will keep running until we break out of it.
        user_input = input("Type 'quit' to exit or press Enter to add a new expense: ")
        
        # Check if the user wants to quit at the start.
        if user_input.lower() == 'quit':
            print("Quitting Program!")
            break  # This break will exit the while loop and end the function.

        # Gather expense details from the user.
        emp_id = input("Enter Employee ID: ")
        if emp_id.lower() == 'quit':
            print("Quitting Program!")
            break  # Exit if 'quit' is entered for Employee ID.
        
        category = input("Enter Expense Category: ")
        if category.lower() == 'quit':
            print("Quitting Program!")
            break  # Exit if 'quit' is entered for Expense Category.

        # Proceed to ask for the expense amount.
        while True:
            amount_input = input("Enter Expense Amount: ")
            if amount_input.lower() == 'quit':
                print("Quitting Program!")
                return  # Use return to exit the function from within this nested loop.
            try:
                amount = float(amount_input)
                if amount < 0:
                    raise ValueError  # Ensure the amount is a positive number.
                break  # Break out of the inner loop once a valid amount is entered.
            except ValueError:
                print("Invalid amount. Please enter a positive number.")

        # Finally, ask for the date of the expense.
        while True:
            date = input("Enter Date (YYYY-MM-DD): ")
            if date.lower() == 'quit':
                print("Quitting Program!")
                return  # Exit the function if 'quit' is entered for the date.
            if is_valid_date(date):
                break  # If the date is valid, break out of the date input loop.
            else:
                print("Invalid date. Please enter the date in YYYY-MM-DD format.")

        # If all inputs are valid, create a new expense record.
        new_expense = {
            "Employee ID": emp_id,
            "Expense Category": category,
            "Amount": amount,
            "Date": date
        }
        expenses.append(new_expense)  # Add the new expense to the list of expenses.

# Assuming expenses is a list that has been previously defined and loaded.
add_expense(expenses)
```

# Remove Unwanted/Incorrect Entires

This section of code is to remove any entries the user would like. It accepts a user's input and removes the unwanted entry

```python
# Define a variable that accepts an input for an employee ID and the date of the expense you would like to remove
def remove_expense(expenses):
    emp_id = input("Enter Employee ID of the expense to remove: ") # Accept input of employee id
    date = input("Enter Date (YYYY-MM-DD) of the expense to remove: ") # Accept input of date

    # Search for the expense in the csv
    found_expense = None
    for expense in expenses:
        if expense["Employee ID"] == emp_id and expense["Date"] == date:
            found_expense = expense
            break

    # If the expense is found, remove it; otherwise, print a message
    if found_expense:
        expenses.remove(found_expense) # Remove item
        print("Expense removed successfully.")
    else:
        print("Expense not found.") # If the item to be removed is not found
```

# Data Analysis and Reporting 

Here we perform some data analysis to get necessary variables for understanding the data. We see expenses broken down by employee, catergory, and then the totals as a whole. This will also be referenced in the visuals below.

```python
# This function calculates the total expenses from all records.
def calculate_total_expenses(expenses):
    """
    Calculate the total amount of expenses.
    Here, we iterate through each expense record and sum up the 'Amount'.
    """
    total = sum(expense["Amount"] for expense in expenses)
    return total

# Calculate total expenses
total_expenses = calculate_total_expenses(expenses)
print(f"Total Expenses: ${total_expenses:.2f}")

# Expenses by Category 
# This function aggregates expenses by their category

# This function calculates the expenses by category
def expenses_by_category(expenses):
    """
    Calculate the total expenses in each category.
    We use a dictionary to keep track of the total expenses per category.
    """
    category_totals = {}
    for expense in expenses:
        category = expense["Expense Category"]
        category_totals[category] = category_totals.get(category, 0) + expense["Amount"]
    return category_totals

# Run the categories total that was defined above
category_totals = expenses_by_category(expenses)
for category, total in category_totals.items():
    print(f"{category}: ${total:.2f}")


# Expenses per Employee
# This function calculates the total expenses for each employee.

def expenses_per_employee(expenses):
    """
    Calculate the total expenses incurred by each employee.
    Similar to the category function, we use a dictionary to sum expenses per employee.
    """
    employee_totals = {}
    for expense in expenses:
        employee_id = expense["Employee ID"]
        employee_totals[employee_id] = employee_totals.get(employee_id, 0) + expense["Amount"]
    return employee_totals

# Run the employees total that was defined above
employee_totals = expenses_per_employee(expenses)
for employee, total in employee_totals.items():
    print(f"Employee {employee}: ${total:.2f}")

```

# Visualization 

Below is a number of visualizations based on our program above that we believe could be used by an employeer to get an easy snapshot of current employee spending. 

# Total Expenses By Category

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define a bar graph of the expenses by category 
def plot_expenses_by_category(expenses):
    """
    Create a bar chart showing expenses by category.
    This visualization helps in quickly understanding which categories have the highest expenses.
    """
    # This generates the expenses by category that will be references in the plot
    categories = expenses_by_category(expenses)
    category_names = list(categories.keys())
    amounts = list(categories.values())

    # Define the chart as a barplot and provide necessary labels
    sns.barplot(x=category_names, y=amounts)
    plt.xlabel('Expense Category')
    plt.ylabel('Total Amount')
    plt.title('Expenses by Category')
    plt.xticks(rotation=45)
    plt.show()

# Display the bar graph
plot_expenses_by_category(expenses)
```

# Total Expenses of Each Employee

```python
# Define the bar graph for employee expenses
def plot_expenses_by_employee(expenses):
    """
    Create a bar chart showing total expenses by employee.
    This visualization helps in understanding the distribution of expenses across employees.
    """
    # Calculate the total expenses by employee ID to be displayed in the chart
    employee_totals = expenses_by_employee(expenses)
    employee_ids = list(employee_totals.keys())
    total_expenses = list(employee_totals.values())

    # Define the type of chart as a barplot and provide labels
    sns.barplot(x=employee_ids, y=total_expenses)
    plt.xlabel('Employee ID')
    plt.ylabel('Total Expenses')
    plt.title('Total Expenses by Employee')
    plt.xticks(rotation=45)
    plt.show()

# Function to calculate total expenses by employee
def expenses_by_employee(expenses):
    employee_totals = {}
    for expense in expenses:
        emp_id = expense["Employee ID"]
        amount = expense["Amount"]
        if emp_id in employee_totals:
            employee_totals[emp_id] += amount
        else:
            employee_totals[emp_id] = amount
    return employee_totals

# View the bar chart
plot_expenses_by_employee(expenses)
```

# Expenses Over Time

```python
import pandas as pd
import matplotlib.dates as mdates

# Define the line chart for expenses over time
def plot_expenses_over_time(expenses):
    """
    Create a line chart showing the trend of expenses over time.
    Each point in the line chart represents total expenses on a particular date.
    """
    
    # Convert dates to datetime objects and sort
    for expense in expenses:
        expense["Date"] = pd.to_datetime(expense["Date"])
    expenses.sort(key=lambda x: x["Date"]) 
    # This allows the dates to be read in to the chart in a consistent format

    # Create a dictionary for us to assign the expenses by date for, this is the line you see on the chart
    expenses_by_date = {}
    for expense in expenses:
        date = expense["Date"]
        expenses_by_date[date] = expenses_by_date.get(date, 0) + expense["Amount"]
    
    dates = list(expenses_by_date.keys())
    amounts = list(expenses_by_date.values())

    # Here we assign the names of our axises and the title of our chart
    sns.lineplot(x=dates, y=amounts)
    plt.xlabel('Date')
    plt.ylabel('Total Amount')
    plt.title('Expenses Over Time')

    # Format the x-axis to display dates in a more readable way
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))  # Adjust interval as needed
    plt.xticks(rotation=45)

    plt.show()
    # The purpose of the above code is to once again format the dates but this time for the x-axis specifically to make them readable to the user

# Plot the graph as we have defined above
plot_expenses_over_time(expenses)
```

# Pie Chart for Expense Distribution 

```python
# Define the expense distribution plot (pie chart)
def plot_expense_distribution(expenses):
    """
    Create a pie chart to show the distribution of expenses across different categories.
    Each slice of the pie chart represents the proportion of expenses in a particular category.
    """
    categories = expenses_by_category(expenses)
    
    # Defining the chart as a pie chart with the three categories taking up the "slices" of the pie
    plt.pie(categories.values(), labels=categories.keys(), autopct='%1.1f%%', startangle=140)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.title('Expense Distribution by Category')
    plt.show()

# Display the pie chart below
plot_expense_distribution(expenses)
```
